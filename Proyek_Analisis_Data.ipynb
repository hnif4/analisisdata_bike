```python
# Import library yang diperlukan
import pandas as pd

# Membaca file CSV ke dalam DataFrame
day_df = pd.read_csv('day.csv')
hour_df = pd.read_csv('hour.csv')

# Menampilkan informasi umum dari setiap DataFrame
print("Informasi Day DataFrame:")
print(day_df.info())
print("\nInformasi Hour DataFrame:")
print(hour_df.info())

# Menampilkan statistik deskriptif dari setiap DataFrame
print("\nStatistik Deskriptif Day DataFrame:")
print(day_df.describe())
print("\nStatistik Deskriptif Hour DataFrame:")
print(hour_df.describe())

# Memeriksa nilai yang hilang di setiap kolom
print("\nNilai yang Hilang di Day DataFrame:")
print(day_df.isnull().sum())
print("\nNilai yang Hilang di Hour DataFrame:")
print(hour_df.isnull().sum())

# Memeriksa duplikasi
print("\nJumlah Baris Duplikat di Day DataFrame:")
print(day_df.duplicated().sum())
print("\nJumlah Baris Duplikat di Hour DataFrame:")
print(hour_df.duplicated().sum())

# Menampilkan beberapa baris pertama dari setiap DataFrame
print("\nBeberapa Baris Pertama Day DataFrame:")
print(day_df.head())
print("\nBeberapa Baris Pertama Hour DataFrame:")
print(hour_df.head())

# Menampilkan beberapa baris terakhir dari setiap DataFrame
print("\nBeberapa Baris Terakhir Day DataFrame:")
print(day_df.tail())
print("\nBeberapa Baris Terakhir Hour DataFrame:")
print(hour_df.tail())
```

### Cleaning Data

```python
# Membaca dataset ke dalam DataFrame
day_df = pd.read_csv('day.csv')
hour_df = pd.read_csv('hour.csv')

# Menampilkan beberapa baris pertama dari setiap dataset
print("Day DataFrame:")
print(day_df.head())
print("\nHour DataFrame:")
print(hour_df.head())

# Cleaning Data
# Menghapus kolom yang tidak diperlukan
day_df_cleaned = day_df.drop(columns=['instant'])
hour_df_cleaned = hour_df.drop(columns=['instant'])

# Mengatasi missing values
# Menampilkan jumlah missing values di setiap kolom
print("\nMissing values in Day DataFrame:")
print(day_df_cleaned.isnull().sum())

print("\nMissing values in Hour DataFrame:")
print(hour_df_cleaned.isnull().sum())

# Menghapus baris dengan missing values (jika ada)
day_df_cleaned = day_df_cleaned.dropna()
hour_df_cleaned = hour_df_cleaned.dropna()

# Menampilkan data yang telah dibersihkan
print("\nCleaned Day DataFrame:")
print(day_df_cleaned.head())

print("\nCleaned Hour DataFrame:")
print(hour_df_cleaned.head())

# Menyimpan data yang telah dibersihkan ke file baru
day_df = pd.read_csv('day.csv')
hour_df = pd.read_csv('hour.csv')

print("\nCleaned data saved successfully.")
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
import pandas as pd
import matplotlib.pyplot as plt

# Pastikan kolom 'dteday' sudah dalam format datetime
day_df['dteday'] = pd.to_datetime(day_df['dteday'])

# Kelompokkan data berdasarkan bulan dan jumlahkan hanya kolom numerik
df_by_month = day_df.groupby(day_df['dteday'].dt.to_period('M')).agg({'cnt': 'sum'})

# Mengubah indeks PeriodIndex menjadi Timestamp
df_by_month.index = df_by_month.index.to_timestamp()

# Membuat subplot dengan ukuran yang telah ditentukan
fig, ax = plt.subplots(figsize=(16, 8))

# Plot garis chart
ax.plot(df_by_month.index, df_by_month['cnt'], label='Jumlah Pengguna Rental Sepeda', marker='o', color='blue', linestyle='-')

# Judul plot dan label sumbu
plt.title('Jumlah Pengguna Rental Sepeda per Bulan selama 2011 - 2012', fontsize=16)
plt.xlabel('Bulan', fontsize=14)
plt.ylabel('Jumlah Pengguna', fontsize=14)

# Menambahkan teks hover untuk setiap titik
for x, y in zip(df_by_month.index, df_by_month['cnt']):
    ax.text(x, y, f'{int(y)}', ha='center', va='bottom')

# Menampilkan legenda
ax.legend()

# Menampilkan grid
plt.grid(True)

# Mengatur layout agar tidak ada yang terpotong
plt.tight_layout()

# Menampilkan plot
plt.show()
```

### Pertanyaan 2:
